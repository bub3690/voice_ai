```python
#같은 환자 찾는 코드
from glob import glob
import shutil

phrase = glob('D:/projects/voice_pathology_ai/voice_data/healthy_new/phrase/export/*-phrase.wav')
a = glob('D:/projects/voice_pathology_ai/voice_data/healthy_new/a/export/*-a_n.wav')
i = glob('D:/projects/voice_pathology_ai/voice_data/healthy_new/i/export/*-i_n.wav')
u = glob('D:/projects/voice_pathology_ai/voice_data/healthy_new/u/export/*-u_n.wav')


#phrase로 dict를 만들고 그 안에 path를 담은 리스트를 넣기.
healthy_dict = {}
for text in phrase:
    text_parse = text.split('\\')[-1].split('-')[0]
    healthy_dict[text_parse]= [text,]

for text in a:
    text_parse = text.split('\\')[-1].split('-')[0]
    try:
        healthy_dict[text_parse].append(text)
    except:
        print(text_parse,"key error")

for text in i:
    text_parse = text.split('\\')[-1].split('-')[0]
    try:
        healthy_dict[text_parse].append(text)
    except:
        print(text_parse,"key error")

        
for text in u:
    text_parse = text.split('\\')[-1].split('-')[0]
    try:
        healthy_dict[text_parse].append(text)
    except:
        print(text_parse,"key error")
        

print(len(healthy_dict))
```
